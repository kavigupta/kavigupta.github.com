```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
rng = np.random.RandomState(100)
```

```python
n = 10000
uncorrelated = rng.randn(2, n)
```

```python
def best_points(data):
    return np.array([point for point in data.T if (data.T <= point).any(axis=1).all()]).T
```

```python
plt.scatter(*uncorrelated)
plt.scatter(*best_points(uncorrelated))
```

```python
u = 1 / np.sqrt(2) * np.array([[1, -1], [1, 1]])
s = np.diag([0.5, 1])
anticorrelated = u @ s @ u.T @ uncorrelated
```

```python
plt.scatter(*anticorrelated)
plt.scatter(*best_points(anticorrelated))
```
