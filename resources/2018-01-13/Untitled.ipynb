```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, "../gerrymandering_data/")
from gerrymandering import *
```

```python
%matplotlib inline
```

```python
def spread(timings):
    return 100 * (np.max(timings, axis=1) - np.min(timings, axis=1))
```

```python
for years in (2004, 2006, 2008, 2010), (2012, 2014, 2016):
    scatterplot(years[-1],
                NamedData(spread(get_all_years(efficiency_gap, years)),
                                "Efficiency Gap Range %s-%s" % (years[0], years[-1])),
                NamedData(spread(get_all_years(effectiveness_gap, years)),
                                "Effectiveness Gap Range %s-%s" % (years[0], years[-1])),
               path="img/range_gaps_%s-%s.png" % (years[0], years[-1]))
```

```python
plt.scatter(spread(get_all_years(efficiency_gap, (2004, 2006, 2008, 2010))),
         spread(get_all_years(effectiveness_gap, (2004, 2006, 2008, 2010))))
add_identity(plt.gca())
```
